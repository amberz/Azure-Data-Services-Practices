Get existing workspace

```python
from azureml.core import Workspace

# get worksapce
ws = Workspace.from_config()
```

Create a new experiment

Note: If the experiment is not found in the workspace, a new experiment is created.

```python
from azureml.core.experiment import Experiment
MyExperiment = Experiment(workspace=ws, name='MyExperiment906')
```

List experiement

```python
print("this workspace has below experiments")
list_MyExperiments = Experiment.list(ws)

for ForMyexperiment in list_MyExperiments:
    print( ForMyexperiment.name)
```

Archive unused experiment

```python
experiment = Experiment(workspace=ws, name="taxi-experiment")
experiment.archive()
```
